```python
#Question 4
#Write a program in python to separate odd and even integers in separate arrays. Go
#to the editor
#Test Data :
#Input the number of elements to be stored in the array :5
#Input 5 elements in the array :
#element - 0 : 25
#element - 1 : 47
#element - 2 : 42
#element - 3 : 56
#element - 4 : 32
#Expected Output:
#The Even elements are:
#42 56 32
#The Odd elements are :
#25 47




def separate_odd_even(arr):
    even_numbers = []
    odd_numbers = []
    for num in arr:
        if num % 2 == 0:
            even_numbers.append(num)
        else:
            odd_numbers.append(num)
    return even_numbers, odd_numbers


num_elements = int(input("Input the number of elements to be stored in the array: "))
array = []
for i in range(num_elements):
    array.append(int(input(f"Element - {i}: ")))


even_elements, odd_elements = separate_odd_even(array)


print("The Even elements are:", end=" ")
for even_num in even_elements:
    print(even_num, end=" ")
print()
print("The Odd elements are:", end=" ")
for odd_num in odd_elements:
    print(odd_num, end=" ")
```

```python
#Question 5
#a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False
#depending on whether the point (x,y) lies in the rectangle with lower left
#corner (x1,y1) and upper right corner (x2,y2).
#>>> inside(1,1,0,0,2,3)
#True
#>>> inside(-1,-1,0,0,2,3)
#False



def inside(x, y, x1, y1, x2, y2):
    
    if (x1 <= x <= x2) and (y1 <= y <= y2):
        return True
    else:
        return False


print(inside(1, 1, 0, 0, 2, 3))  
print(inside(-1, -1, 0, 0, 2, 3))  

#b. Use function inside() from part a. to write an expression that tests whether
#the point (1,1) lies in both of the following rectangles: one with lower left
#corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower
#left corner (0.5, 0.2) and upper right corner (1.1, 2)



def inside(x, y, x1, y1, x2, y2):
   
    if (x >= x1 and x <= x2) and (y >= y1 and y <= y2):
        return True
    else:
        return False

point = (1, 1)
rect1 = (0.3, 0.5, 1.1, 0.7)
rect2 = (0.5, 0.2, 1.1, 2)

in_rect1 = inside(point[0], point[1], rect1[0], rect1[1], rect1[2], rect1[3])
in_rect2 = inside(point[0], point[1], rect2[0], rect2[1], rect2[2], rect2[3])

if in_rect1 and in_rect2:
    print("The point", point, "lies in both rectangles.")
else:
    print("The point", point, "does not lie in both rectangles.")
```

```python
#Question 6
#16. You can turn a word into pig-Latin using the following two rules (simplified):
#• If the word starts with a consonant, move that letter to the end and append
#'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
#• If the word starts with a vowel, simply append 'way' to the end of the word.
#For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For
#our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
#Write a function pig() that takes a word (i.e., a string) as input and returns its pig-
#Latin form. Your function should still work if the input word contains upper case
#characters. Your output should always be lower case however.
#>>> pig('happy')
#'appyhay'
#>>> pig('Enter')
#'enterway'



def pig(word):
   
    vowels = ['a', 'e', 'i', 'o', 'u']
    word = word.lower() 
    if word[0] in vowels:
        return word + 'way'
    else:
        return word[1:] + word[0] + 'ay'


print(pig('happy'))  
print(pig('Enter')) 
```

```python
#Question 7
#File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.
#Write a function bldcount() that reads the file with name name and reports (i.e.,
#prints) how many patients there are in each bloodtype.
#>>> bldcount('bloodtype.txt')
#There are 10 patients of blood type A.
#There is one patient of blood type B.
#There are 10 patients of blood type AB.
#There are 12 patients of blood type O.
#There are no patients of blood type OO.



def bldcount(filename):
   
    blood_count = {'A': 0, 'B': 0, 'AB': 0, 'O': 0, 'OO': 0}
    
    
    with open(filename, 'r') as file:
        for line in file:
            blood_type = line.strip()  
            blood_count[blood_type] += 1  
    
 
    for blood_type, count in blood_count.items():
        if count == 1:
            print("There is one patient of blood type {}.".format(blood_type))
        elif count > 1:
            print("There are {} patients of blood type {}.".format(count, blood_type))
        else:
            print("There are no patients of blood type {}.".format(blood_type))




```
